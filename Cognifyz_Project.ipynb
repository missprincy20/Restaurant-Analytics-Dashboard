Step 1: Load Dataset

```python
import streamlit as st
import pandas as pd

st.set_page_config(
    page_title="Restaurant Analytics Dashboard",
    layout="wide"
)

@st.cache_data
def load_data():
    df = pd.read_csv("/content/sample_data/Dataset .csv", encoding="latin1")
    return df

df = load_data()

st.title("🍽 Restaurant Analytics Dashboard")
```

Step 2: Sidebar Filters

```python
st.sidebar.header("Filters")

city = st.sidebar.selectbox(
    "Select City",
    sorted(df["City"].dropna().unique())
)

cuisine = st.sidebar.selectbox(
    "Select Cuisine",
    sorted(df["Cuisines"].dropna().unique())
)

price = st.sidebar.selectbox(
    "Price Range",
    sorted(df["Price range"].unique())
)

rating = st.sidebar.slider(
    "Minimum Rating",
    0.0,
    5.0,
    3.0
)

delivery = st.sidebar.selectbox(
    "Online Delivery",
    ["All","Yes","No"]
)
```

Step 3: Apply Filters

```python
filtered = df.copy()

filtered = filtered[
    filtered["City"] == city
]

filtered = filtered[
    filtered["Aggregate rating"] >= rating
]

filtered = filtered[
    filtered["Price range"] == price
]

filtered = filtered[
    filtered["Cuisines"] == cuisine
]

if delivery != "All":
    filtered = filtered[
        filtered["Has Online delivery"] == delivery
    ]
```

Step 4: KPIs

```python
col1, col2, col3, col4 = st.columns(4)

if filtered.empty:
    col1.metric("Restaurants", 0)
    col2.metric("Average Rating", 0)
    col3.metric("Average Votes", 0)
    col4.metric("Average Cost", 0)
else:
    col1.metric(
        "Restaurants",
        len(filtered)
    )

    col2.metric(
        "Average Rating",
        round(filtered["Aggregate rating"].mean(), 2)
    )

    col3.metric(
        "Average Votes",
        int(filtered["Votes"].mean())
    )

    col4.metric(
        "Average Cost",
        round(filtered["Average Cost for two"].mean(), 2)
    )
```

Step 5: Restaurant Table

```python
st.subheader("Restaurants")

st.dataframe(
    filtered[
        [
            "Restaurant Name",
            "Aggregate rating",
            "Votes",
            "Price range",
            "Cuisines"
        ]
    ]
)
```

Step 6: Interactive Rating Chart

```python
import plotly.express as px

fig = px.histogram(
    filtered,
    x="Aggregate rating",
    nbins=20,
    title="Rating Distribution"
)

st.plotly_chart(fig,use_container_width=True)
```

Step 7: Cuisine Chart

```python
if filtered.empty:
    st.warning("No data available for selected filters.")
else:
    top = (
        filtered["Cuisines"]
        .value_counts()
        .head(10)
        .reset_index()
    )

    top.columns = ["Cuisine", "Count"]

    fig = px.bar(
        top,
        x="Count",
        y="Cuisine",
        orientation="h",
        title="Top Cuisines"
    )

    st.plotly_chart(fig, use_container_width=True)
```

Step 8: Map

```python
fig = px.scatter_map(
    filtered,
    lat="Latitude",
    lon="Longitude",
    hover_name="Restaurant Name",
    hover_data=[
        "Aggregate rating",
        "Votes"
    ],
    zoom=10
)

st.plotly_chart(fig,use_container_width=True)
```

Feature 4: Restaurant Recommendation System

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer()

matrix = tfidf.fit_transform(
    df["Cuisines"].fillna("")
)
```

```python
similarity = cosine_similarity(matrix)
```

```python
restaurant = st.selectbox(
    "Choose Restaurant",
    df["Restaurant Name"].unique()
)
```

```python
#Recommendation Function

index = df[
    df["Restaurant Name"]==restaurant
].index[0]

scores = list(
    enumerate(similarity[index])
)

scores = sorted(
    scores,
    key=lambda x:x[1],
    reverse=True
)

recommended = []

for i in scores[1:6]:

    recommended.append(
        df.iloc[i[0]]["Restaurant Name"]
    )

st.write(recommended)
```
